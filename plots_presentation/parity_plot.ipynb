```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
y_true = np.linspace(0, 100, 100)  # True values
noise = np.random.normal(0, 5, size=y_true.shape)  # Random noise
y_pred = y_true + noise  # Predicted values with added noise

# Create parity plot
plt.figure(dpi=200, figsize=(8, 8))
plt.scatter(y_true, y_pred, alpha=0.7, edgecolor='k', color='skyblue', label='Predictions with Noise')

# Plot the ideal parity line (perfect prediction)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2, label='Perfect Prediction')

# Labels and title
plt.xlabel(r'Calculated Value', fontsize=14)
plt.ylabel(r'Predicted Value', fontsize=14)
plt.title('Parity Plot', fontsize=16)

plt.grid(True)

# Adjust layout for clean look
plt.tight_layout()
plt.show()
```
