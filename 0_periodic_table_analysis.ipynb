```python
import pandas as pd
from tqdm import tqdm

df1 = pd.read_csv('./Data/1_MatDX/MatDX_nomad_EF_Binary.csv')
df2 = pd.read_csv('./Data/1_MatDX/MatDX_nomad_EF_Ternary.csv')

# Concatenate the DataFrames
df = pd.concat([df1, df2], ignore_index=True)
```

```python
element_dict = {}

for index, row in tqdm(df.iterrows(), total=len(df), desc="Processing Rows"):
    for element in eval(row['formation_energy'])['reference'].keys():
        if element in element_dict.keys():
            element_dict[element] += 1
        else:
            element_dict[element] = 1
```

```python
element_dict
```

```python
df = pd.DataFrame(list(element_dict.items()), columns=['Element', 'Value'])
```

```python
df.to_csv('element.csv', index=False)
```
